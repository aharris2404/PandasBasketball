```python
from PandasBasketball import pandasbasketball as pb
import pandas as pd
from bs4 import BeautifulSoup as soup
import seaborn as sns
```

```python
df = pb.get_player("jamesle01", "playoffs_advanced")
df.head()
```

```python
for column_name in df.columns:
    if column_name not in ['Season', 'PER']:
        df = df.drop([column_name], axis=1)
        
df['avg_per'] = [15.0 for i in range(13)]
df['PER'] = df['PER'].astype(float)
df['Season'] = [int(season[-2:]) for season in df['Season']]

df.info()
```

```python
fig, ax = sns.lineplot(x=df['Season'], y=df['PER'])
```
